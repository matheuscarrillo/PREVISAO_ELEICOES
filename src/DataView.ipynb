```python
def gera_base_aux(coluna_req):
    colunas_requeridas = ['date']
    for coluna in df_twitter.columns:
        if coluna_req in coluna:
            colunas_requeridas.append(coluna)
    df_aux = df_twitter[colunas_requeridas]
    df_aux['anomes']=df_aux['date'].apply(lambda row: row.split('-')[0]+'-'+row.split('-')[1])
    df_aux['date'] = pd.to_datetime(df_aux['date'], format='%Y-%m-%d')
    return df_aux
```

```python
coluna_req = 'tweet_count'
df_tweet = gera_base_aux('tweet_count')
```

```python
df_groupby_tweet = df_tweet.groupby(['anomes']).sum()
```

#### Avaliando correlação entre a contagem de tweets

```python
correlations = df_groupby_tweet.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(correlations, vmin=-1, vmax=1, annot=True)
plt.show()
```

Conclusões (validar na base completa):

- Tweets do lula estão mais relacionados a saúde e eleições

- Quando lula fala de saúde, tende a falar de economia

- Quando lula cita eleições, tende a falar sobre outros assuntos

- Tweets de bolsonaro falam mais de saúde e economia

#### Medindo engajamento

(1000 * ((reply + retweet + favorite)/(contagem de twitter)) / (num_seguidores))

```python
#### Fonte: https://buzzmonitor.com.br/blog/como-o-engajamento-e-calculado-para-cada-rede-social/#:~:text=M%C3%A9dia%3A%20M%C3%A9dia%20do%20engajamento%20%C3%A9%20feita%20da%20seguinte%20forma%3A%20Dividimos,e%20ent%C3%A3o%20multiplicado%20por%201.000.
```

```python
df_twitter['engajamento_[geral]_Bolsonaro']= (df_twitter['reply_mean_[geral]_Bolsonaro']+df_twitter['retweet_mean_[geral]_Bolsonaro']+df_twitter['favorite_mean_[geral]_Bolsonaro'])/df_twitter['tweet_count_[geral]_Bolsonaro']
df_twitter['engajamento_[geral]_Lula']= (df_twitter['reply_mean_[geral]_Lula']+df_twitter['retweet_mean_[geral]_Lula']+df_twitter['favorite_mean_[geral]_Lula'])/df_twitter['tweet_count_[geral]_Lula']
```

```python
df_twitter['date'] = pd.to_datetime(df_twitter['date'])
```

```python
df_twitter['anomes']=df_twitter['date'].apply(lambda row: row.strftime('%Y-%m'))
```

```python
df_twitter_agrupada_anomes = df_twitter.groupby(['anomes']).mean().reset_index()
```

```python
df_twitter_agrupada_anomes = df_twitter_agrupada_anomes[['anomes', 'engajamento_[geral]_Bolsonaro', 'engajamento_[geral]_Lula']]
```

```python
df_engajamento = pd.merge(df_twitter_agrupada_anomes, df_seguidores, how='left', on='anomes')
```

```python
df_engajamento['engajamento_[geral]_Bolsonaro']=1000*(df_engajamento['engajamento_[geral]_Bolsonaro']/df_engajamento['Bolsonaro'])
df_engajamento['engajamento_[geral]_Lula']=1000*(df_engajamento['engajamento_[geral]_Lula']/df_engajamento['Lula'])
```

```python
plt.figure(figsize=(20,10))
plt.plot('anomes', 'engajamento_[geral]_Bolsonaro', data=df_engajamento)
plt.plot('anomes', 'engajamento_[geral]_Lula', data=df_engajamento)

plt.title("Engajamento ao longo dos meses",fontsize=15)
plt.xlabel("Ano-Mês",fontsize=13)
plt.ylabel("Engajamento",fontsize=13)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
